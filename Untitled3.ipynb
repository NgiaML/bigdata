```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

p="/content/sds.csv"
df = pd.read_csv(p)
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Ngày']=pd.to_datetime(df.Ngày)
df.drop_duplicates(inplace=True)
print(df)
```

```python
df['month']=df.Ngày.dt.month
df['date']=df.Ngày.dt.day
print(df)
```

```python
tongsoca=df.groupby(by=['Tỉnh']).sum(numeric_only=True)['Tổng số ca lây nhiễm']
max=tongsoca.max()
print(tongsoca)
```

```python
for Tỉnh, soca in tongsoca.items():
    if soca==max:
        print(f'\n Tỉnh (TP) {Tỉnh} có số ca nhiều nhất: {soca} ca nhiễm')
```

```python
tinhtp=[]
for tinh, soca in tongsoca.items():
  tinhtp.append(tinh)

plt.figure(figsize=[24,14])
plt.bar(tinhtp, tongsoca)
plt.title("BIỂU ĐỒ TỔNG SỐ CA CỦA CÁC TỈNH TỪ NGÀY 06/08 ĐẾN NGÀY 29/08  ",size=30,pad=15)
# Định dạng giá trị trục x,y trên biểu đồ
plt.xticks(tinhtp,rotation=95,size=15)
plt.yticks(size=16)
#  Định dạng tiêu đề cột x,y 
plt.xlabel("Tỉnh",size=18,labelpad=5)
plt.ylabel("Tổng số ca",size=18)
# Tạo lưới biểu đồ
plt.grid(color='#95a5a6',  axis='y')
plt.show()
```

```python
socatheongay=df.groupby(['month','date']).sum(numeric_only=True)['Tổng số ca lây nhiễm']
print(socatheongay)
max=socatheongay.max()
```

```python
for ngay, soca in socatheongay.items():
    if soca==max:
        print(f'Ngày có tổng số ca cao nhất là: {ngay[1]}/{ngay[0]} với tổng số ca {soca}')
```

```python
date=[]
for ngay, soca in socatheongay.items():
    a=f'{ngay[1]}/{ngay[0]}'
    date.append(a)

plt.figure(figsize=[22,10])

plt.plot(date,socatheongay,c = 'r',marker = 'o')
plt.title("BIỂU ĐỒ TỔNG SỐ CA NHIỄM CỦA CÁC TỈNH TỪ NGÀY 06/08 ĐẾN 29/08  ",size=30,pad=15)

plt.xticks(date,rotation=90,size=16)
plt.yticks(size=16)
plt.xlabel("Tháng",size=18,labelpad=10)

plt.ylabel("Tổng doanh thu",size=18)

plt.grid(color='#95a5a6',  axis='x')
plt.grid(color='#95a5a6',  axis='y')

plt.show()
```
